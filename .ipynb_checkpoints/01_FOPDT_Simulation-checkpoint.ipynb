## 1. FOPDT Simulation for Generating training Data

```python
from gekko import GEKKO
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# from smt.sampling_methods import LHS
# from smt.sampling_methods import Random
```

```python
tf = 30 # fianl time
```

```python
m=GEKKO(remote=False)
m.time = np.linspace(0,tf-1,tf)
```

```python
K = m.FV(1) # Process Gain
tau = m.FV(2) # Time Constant
```

```python
u_input = np.zeros(tf)
u_input[5:] = 1
```

```python
y = m.CV()
u = m.MV(u_input)
m.Equation(tau*y.dt()+y==K*u) #FOPDT Equation

m.options.IMODE=4
```

```python
m.solve(disp=False)
```

```python
plt.figure(0)
plt.subplot(2,1,1)
plt.plot(m.time, y)
plt.subplot(2,1,2)
plt.plot(m.time, u, drawstyle='steps')
plt.show()
```

## 2. Generating input signal (Random and LHS)

```python
import numpy as np
import matplotlib.pyplot as plt

nstep = 400 # Choose training data lenth

# random signal generation

a_range = [0,2]
a = np.random.rand(nstep) * (a_range[1]-a_range[0]) + a_range[0] # range for amplitude
a[0] = 0

b_range = [5, 20]
b = np.random.rand(nstep) *(b_range[1]-b_range[0]) + b_range[0] # range for frequency
b = np.round(b)
b = b.astype(int)

b[0] = 0

for i in range(1,np.size(b)):
    b[i] = b[i-1]+b[i]

# Random Signal
i=0
random_signal = np.zeros(nstep)
while b[i]<np.size(random_signal):
    k = b[i]
    random_signal[k:] = a[i]
    i=i+1

# PRBS
a = np.zeros(nstep)
j = 0
while j < nstep:
    a[j] = 5
    a[j+1] = -5
    j = j+2

i=0
prbs = np.zeros(nstep)
while b[i]<np.size(prbs):
    k = b[i]
    prbs[k:] = a[i]
    i=i+1

plt.figure(0) 
plt.subplot(2,1,1)
plt.plot(random_signal, drawstyle='steps',label='Random Signal')
plt.legend()
plt.subplot(2,1,2)
plt.plot(prbs, drawstyle='steps', label='PRBS')
plt.legend()
plt.show()
```

## 3. Evaluate the Random Input Signal

```python
tf = nstep
m.time = np.linspace(0,tf-1,tf)
u.value = random_signal
```

```python
m.options.IMODE = 4
m.solve(disp=False)
```

```python
plt.figure(0)
plt.subplot(2,1,1)
plt.plot(m.time, y)
plt.subplot(2,1,2)
plt.plot(m.time, u, drawstyle='steps')
plt.show()
```

## 4. LSTM fitting

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import joblib
import time
```

```python
# For LSTM model
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
from keras.callbacks import EarlyStopping
from keras.models import load_model
```

```python
window = 5
#Load training data
data = pd.DataFrame(
        {"u": u,
         "y": y},
        index = np.linspace(1,nstep,nstep,dtype=int))
```

```python
# Scale features
s1 = MinMaxScaler(feature_range=(-1,1))
data_s = s1.fit_transform(data[['u','y']])
```

```python
plt.plot(data_s)
```

```python
joblib.dump(s1, 's1.sav')
```

```python
U = []
Y = []
for i in range(window,len(data_s)):
    U.append(data_s[i-window:i,:])
    Y.append(data_s[i])
```

```python
# Reshape data to format accepted by LSTM
U, Y = np.array(U), np.array(Y)
```

```python
# # Initialize LSTM model
model = Sequential()

model.add(LSTM(units=100, return_sequences=True, \
          input_shape=(U.shape[1],U.shape[2])))
model.add(Dropout(0.2))
model.add(LSTM(units=100, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=100))
model.add(Dropout(0.2))
model.add(Dense(units=8)) #units = number of outputs
model.compile(optimizer = 'adam', loss = 'mean_squared_error',\
              metrics = ['accuracy'])
# Allow for early exit
es = EarlyStopping(monitor='loss',mode='min',verbose=1,patience=10)
```
